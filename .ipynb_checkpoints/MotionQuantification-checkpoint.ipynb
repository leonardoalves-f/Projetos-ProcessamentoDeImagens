```python
%matplotlib inline
import imageio
import numpy as np
import matplotlib.pyplot as plt
import cv2

from skimage.color import rgb2gray

im1=imageio.imread('./frame07.png')
im2=imageio.imread('./frame11.png')

im1 = rgb2gray(im1)
im2 = rgb2gray(im2)

plt.figure(figsize=(15,4))

plt.subplot(121),plt.imshow(im1, cmap = 'gray')
plt.title('Motion sequence 1'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(im2, cmap = 'gray')
plt.title('Motion sequence 2'), plt.xticks([]), plt.yticks([])


plt.show()
```

## Good features to track:
(adapted from https://stackoverflow.com/questions/61035461/to-find-optical-flow-as-ndarray-using-cv2-calcopticalflowpyrlk
"To find optical flow as ndarray using cv2.calcOpticalFlowPyrLK()"

Jianbo Shi and Tomasi, "Good features to track," 1994 Proceedings of IEEE Conference on Computer Vision and Pattern Recognition, Seattle, WA, USA, 1994, pp. 593-600, doi: 10.1109/CVPR.1994.323794.

```python
# params for ShiTomasi corner detection
feature_params = dict( maxCorners = 100,
                       qualityLevel = 0.3,
                       minDistance = 7,
                       blockSize = 7 )


p0 = cv2.goodFeaturesToTrack(im1, mask = None, **feature_params)
```

```python
# Plot selected features over the original image. 
plt.imshow(im1,cmap='gray')
plt.show()

plt.imshow(im1,cmap='gray')
p0 = np.squeeze(p0) # remove singleton dimension from array
xa, ya = p0.T
xa.shape
plt.scatter(xa,ya,c='red')
```

```python

# Parameters for lucas kanade optical flow
lk_params = dict( winSize  = (15,15),
                  maxLevel = 2,  # zero based maximal pyramid level number.  Here, two levels are used.
                  criteria = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 10, 0.03))


p1, st, err = cv2.calcOpticalFlowPyrLK(im1,im2, p0, None, **lk_params)


print(f"p0 {p0.shape}")
print(f"p1 {p1.shape}")
print(f"st {st.shape}")
print(f"xa {xa.shape}")
```

### Observe below that p0 are the corner detections in the original image that we gave to LK.
### p1 are the locations of the corners in the second image, tracked by the algorithm LK.

We can plot the first point from p0 and p1 to see the correspondence:
Corner at [266, 221]  moved to [270.446, 225.998]

There are 77 corner detections in this image.  We can plot them all

First plot will be a scatter of the corners in the first image (red) and then the corners in the second image. 

Second plot will use the matplotlib.pyplot.arrows to plot this same correspondence of points.  We can check that the arrow is 
in the correct direction, etc. 

```python
print(p0[0,:])
print(p1[0,:])

plt.figure(figsize=(16,16))
plt.imshow(im1,cmap='gray')
#p0 = np.squeeze(p0) # remove singleton dimension from array
x0 = p0[:,0]
y0 = p0[:,1]
x1 = p1[:,0]
y1 = p1[:,1]
#xa.shape

npoint, = x0.shape
print(npoint)

plt.scatter(x0,y0,c='red')
plt.scatter(x1,y1,c='blue')
plt.show()

plt.figure(figsize=(16,16))
plt.imshow(im1,cmap='gray')
#p0 = np.squeeze(p0) # remove singleton dimension from array
dx = x1 - x0
dy = y1 - y0
#xa.shape
for i in range(npoint):    
    plt.arrow(x0[i], y0[i], dx[i], dy[i], length_includes_head = True, head_width=2.0, head_length = 0.5, color='red')
plt.show()
```

```python
print(f"Status = {st.T}")  # Output status vector.  Each element of the vector is set to 1 if the flow for the corresponding
# feature has been found, otherwise it is set to zero.
print(f"Error = {err.T}")  # Output vector of errors.  Each element of the vector is set to an error for the corresponding
# feature, type of error measure can be set in flags parameer.  If the flow wasn`t found, then error is not defined. 
```

```python
import cv2

flow = cv2.calcOpticalFlowFarneback(im1,im2, None, 0.5, 3, 15, 3, 5, 0, 0)
#CalcOpticalFlowBM(im1,im2, blockSize, shiftSize, max_range, usePrevious, velx, vely)
print(flow)

u = flow[:,:,0]
v = flow[:,:,1]
r,c = u.shape
print(r,c)
```

```python

x = np.arange(0,u.shape[1])
y = np.arange(0,u.shape[0])
print(x.shape)
print(y.shape)
X, Y = np.meshgrid(x, y)
print(X.shape)

plt.imshow(u)
plt.title('x component of velocity')
plt.colorbar()
plt.show()

plt.imshow(v)
plt.title('y component of velocity')
plt.colorbar()
plt.show()
```
