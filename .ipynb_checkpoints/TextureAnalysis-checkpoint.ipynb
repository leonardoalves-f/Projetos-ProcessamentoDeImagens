# GLCM Texture Features


This example illustrates texture classification using grey level
co-occurrence matrices (GLCMs). A GLCM is a histogram of co-occurring
greyscale values at a given offset over an image.

```python
import matplotlib.pyplot as plt
import numpy as np
from skimage.feature import greycomatrix, greycoprops
I = np.array([[1, 1, 7, 5, 3, 2],
              [5, 1, 6, 1, 2, 5],
              [8, 8, 6, 8, 1, 2],
              [4, 3, 4, 5, 5, 1],
              [8, 7, 8, 7, 6, 2],
              [7, 8, 6, 2, 6, 2]])
print(I)

# We want to reproduce an example given in class from Gonzalez:
glcm = greycomatrix(I, distances=[1], angles=[0], levels= 9, symmetric=False, normed=False)

# We choose distances = 1 and angles = 0 to choose q as the pixel one to the right of p. (i.e pq).
# We use 9 levels, since we need [0,8].  0 is redundant in this case. 
print(glcm)
```

### Plot image:

```python
plt.imshow(I)
plt.title("Original Image, I")
plt.show
```

## Note that our rule is q is pixel to the right of p.  There are 6 rows with pairs. There can be 5 pairs in each row (6 minus 1), so a total of 30 pairs in this image. 

### Plot GLCM
glcm is of dimension (8x8x1x1), with two singleton dimensions.  Must use squeeze to transform to 8x8 matrix.

```python
print(f'Number of dimensions in glcm matrix: {np.shape(glcm)}')
glcm = np.squeeze(glcm)
a,b=np.shape(glcm)
print(f'Removed singleton dimensions: {np.shape(glcm)}')
print(glcm)

plt.imshow(glcm)
plt.title("Graylevel Cooccurrence Matrix")
plt.show()

print(f'Min value of glcm is {np.min(glcm)}, Max value of glcm is {np.max(glcm)}')
```

## Notice that glcm is not, in general, symmetric.  
## for a pair of pixels, pq (e.g. 6,2) the count is NOT the same as for (2,6).

### Plot same glcm matrix, but this time with numbers.

## The GLCM can become very large. We want to "bin" the grayscale level to reduce the size of the GLCM.
## Imagine the original image has levels 1 to 6.  We map {1,2}$\rightarrow$ 1, {3,4}$\rightarrow$ 2, {5,6} $\rightarrow$ 3.
## We reduce grayscale level by two, and the size of the matrix by 4. 

## (a+1)//2  uses Integer division by 2.  This accomplishes our mapping.

## 1d example:

```python
a = np.array([1,2,3,4,5,6])
r = np.shape(a)
b=np.zeros(r)
print(a)
b = (a+1)//2  # Integer division by 2
print(b)
```

## 2d example:

```python
I = np.array([[1, 1, 7, 5, 3, 2],
              [5, 1, 6, 1, 2, 5],
              [8, 8, 6, 8, 1, 2],
              [4, 3, 4, 5, 5, 1],
              [8, 7, 8, 7, 6, 2],
              [7, 8, 6, 2, 6, 2]])
print(f'Image, I = \n{I}')
r,c = np.shape(I)

# Integer division by 2
I2 =(I+1)//2
print(f'Image, I2 = \n{I2}')
```

```python
glcm = greycomatrix(I2, distances=[1], angles=[0], levels= 5, symmetric=False, normed=False)

#Remove singleton dimensions
glcm = np.squeeze(glcm)
a,b=np.shape(glcm)
print(f'Removed singleton dimensions: {np.shape(glcm)}')
print(glcm)

plt.imshow(glcm)
plt.title("Graylevel Cooccurrence Matrix")
plt.show()

print(f'Min value of glcm is {np.min(glcm)}, Max value of glcm is {np.max(glcm)}')
```

### Sum number of values in glcm matrix

```python
sumglcm = np.sum(glcm)
print(sumglcm)
```

### Now create a new matrix, P, which is contains the probabilities.

```python
P = glcm/sumglcm
print(P)
```

### Extract features of GLCM

```python
from skimage.feature import greycomatrix, greycoprops
from skimage import data
# We are going to store the properties in xs and ys.

glcmStack = [glcm1,glcm2,glcm3] #Put our squeezed glcms for the 3 images into a 3D matrix.

vs = []
ws = []
xs = []
ys = []
zs = []

for glcmSlice in glcmStack: # We iterate over the first dimension of glcmStack (i.e. 0 ate 2)
    vs.append(greycoprops(glcmSlice, prop='correlation'))
    ws.append(greycoprops(glcmSlice, prop='contrast'))
    xs.append(greycoprops(glcmSlice, prop='energy'))
    ys.append(greycoprops(glcmSlice, prop='homogeneity'))
    zs.append(greycoprops(glcmSlice, prop='dissimilarity'))
    
# print values

for i in range(0,3):
    print(f"Correlation = {vs[i]}, Contrast = {ws[i]}, Energy Correlation = {xs[i]}", 
          f", Homogeneity = {ys[i]}, Dissimilarity = {zs[i]} \n")
```

```python
# compute some GLCM properties each patch
glcmSqStack = [glcmSq1,glcmSq2,glcmSq3] #Put our squeezed glcms for the 3 images into a 3D matrix.
print(f'Shape of glcmslice (dimensions): {np.shape(glcmStack)}')  # glcmStack is of size

# We iterate over the first dimension of glcmStack (i.e. 0 ate 2).
# Also, 'enumerate' lets us get both an index (i.e. number 0 to 2 in this case) AND the slice of matrix.

for idx,glcmSqSlice in enumerate(glcmSqStack): 
    print(f'glcmslice {idx}: \n {glcmSqSlice[0:3,0:3]}') #print a small portion of each matrix
    
```

## Entropia
### Algoritmo não disponível no sci-kit.
### Measures the randomness of the elements of matrix G. Entropy = 0 when all $p_{ij}$ are 0 and maximum when $p_{ij}$ are equal
### Max value is $2log_2K$.  

# $H=-\sum_{i=1}^K \sum_{j=1}^K p_{ij}log_2p_{ij}$
#### Gonzalez and Woods 2008


```python
glcmStack = [glcm1,glcm2,glcm3] #Put our  glcms for the 3 images into a 3D matrix.
v,w,x,y,z=np.shape(glcmStack)

for idx,glcmSlice in enumerate(glcmStack): # We iterate over the first dimension of glcmStack (i.e. 0 ate 2)
    sumglcm = np.sum(glcmSlice)
    P = np.squeeze(glcmSlice/sumglcm)  # normalize slice and remove singletons
    H = -P*np.log(P+ np.finfo(float).eps)  # eps is taken to be a very small value, so as not to allow zero values in log function.
    print(sum(sum(H)))
```

# Cameraman example
Adapted from the following site by John Sims 2020.
https://scikit-image.org/docs/dev/auto_examples/features_detection/plot_glcm.html

"In this example, samples of two different textures are extracted from
an image: grassy areas and sky areas. For each patch, a GLCM with
a horizontal offset of 5 is computed. Next, two features of the
GLCM matrices are computed: dissimilarity and correlation. These are
plotted to illustrate that the classes form clusters in feature space.

In a typical classification problem, the final step (not included in
this example) would be to train a classifier, such as logistic
regression, to label image patches from new images."

I did not use logistic regression, but present an example of cluster analysis.


```python
import matplotlib.pyplot as plt
import numpy as np
from skimage.feature import greycomatrix, greycoprops
from skimage import data

# open the camera image
image = data.camera()


# create the figure
fig = plt.figure(figsize=(8, 8))
plt.imshow(image, cmap=plt.cm.gray,
          vmin=0, vmax=255)

```

```python
## Get four patches of grass,  four patches of sky and plot on original image.
```

```python
PATCH_SIZE = 21

# select some patches from grassy areas of the image
grass_locations = [(474, 291), (440, 433), (466, 18), (462, 236)]
grass_patches = []
for loc in grass_locations:
    grass_patches.append(image[loc[0]:loc[0] + PATCH_SIZE,
                               loc[1]:loc[1] + PATCH_SIZE])

# select some patches from sky areas of the image
sky_locations = [(54, 48), (21, 233), (90, 380), (195, 330)]
sky_patches = []
for loc in sky_locations:
    sky_patches.append(image[loc[0]:loc[0] + PATCH_SIZE,
                             loc[1]:loc[1] + PATCH_SIZE])

# create the figure
fig = plt.figure(figsize=(8, 8))

# display original image with locations of patches
ax = fig.add_subplot(1,1, 1)
ax.imshow(image, cmap=plt.cm.gray,
          vmin=0, vmax=255)
greencolor = (0,1,0)
bluecolor = (0,0,1)
for (y, x) in grass_locations:
    rect = plt.Rectangle((x,y),PATCH_SIZE,PATCH_SIZE, fill=False,
            edgecolor=greencolor, linewidth=2.5)
    plt.gca().add_patch(rect)
    #ax.plot(x + PATCH_SIZE , y + PATCH_SIZE , 'gs')
for (y, x) in sky_locations:
    rect = plt.Rectangle((x,y),PATCH_SIZE,PATCH_SIZE, fill=False,
            edgecolor=bluecolor, linewidth=2.5)
    plt.gca().add_patch(rect)
ax.set_xlabel('Original Image')
ax.set_xticks([])
ax.set_yticks([])
ax.axis('image')
plt.show()

```

## grass_patches consists of 4 arrays, each containing a 21x21 region of pixels.

```python
print(np.shape(grass_patches))
```

## Plot each of the patches we have taken as subplots.  Notice that the texture of the grass subplot is similar,  and texture of sky subplots is similar.

```python
fig = plt.figure(figsize=(8, 8))

# display the image patches
for i, patch in enumerate(grass_patches):
    ax = fig.add_subplot(3, len(grass_patches), len(grass_patches)*1 + i + 1)
    ax.imshow(patch, cmap=plt.cm.gray,
              vmin=0, vmax=255)
    ax.set_xlabel('Grass %d' % (i + 1))

for i, patch in enumerate(sky_patches):
    ax = fig.add_subplot(3, len(sky_patches), len(sky_patches)*2 + i + 1)
    ax.imshow(patch, cmap=plt.cm.gray,
              vmin=0, vmax=255)
    ax.set_xlabel('Sky %d' % (i + 1))
```

## For each subplot, we compute gray level co-occurrence matrix and extract the features "dissimilarity" and "correlation". 

```python

# compute some GLCM properties each patch
xs = []
ys = []
for patch in (grass_patches + sky_patches):
    glcm = greycomatrix(patch, [5], [0], 256, symmetric=True, normed=True)
    xs.append(greycoprops(glcm, 'dissimilarity')[0, 0])
    ys.append(greycoprops(glcm, 'correlation')[0, 0])

# print values for dissimilarity and correlation
end_xs, = np.shape(xs)
# grass patches, then sky patches
for i in range(0,end_xs):
    print(f"Dissimilarity = {xs[i]}, Correlation = {ys[i]}\n")

    
```

## Look at dimensions for one of the GLCMs.  

```python
print(np.shape(glcm))
```

## Plot features

```python
    
fig = plt.figure(figsize=(8, 8))

# for each patch, plot (dissimilarity, correlation)
ax = fig.add_subplot(1, 1, 1)
ax.plot(xs[:len(grass_patches)], ys[:len(grass_patches)], 'go',
        label='Grass')
ax.plot(xs[len(grass_patches):], ys[len(grass_patches):], 'bo',
        label='Sky')
ax.set_xlabel('GLCM Dissimilarity')
ax.set_ylabel('GLCM Correlation')
ax.legend()


# display the patches and plot
fig.suptitle('Grey level co-occurrence matrix features', fontsize=14)
plt.show()
```

## Hierarchical clustering

A natural next step might be to determine the textures based on clustering.  Here we use hierarchical clustering

https://scikit-learn.org/stable/modules/clustering.html#hierarchical-clustering
https://stackabuse.com/hierarchical-clustering-with-python-and-scikit-learn/
https://joernhees.de/blog/2015/08/26/scipy-hierarchical-clustering-and-dendrogram-tutorial/

### First put out features, xs = Dissimilarity and ys=Correlation into columns of a matrix. 

```python


r, = np.shape(xs)
X = np.zeros([r,2])

X[:,0] = xs
X[:,1] = ys
print(X)
```

```python


from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt

linked = linkage(X, 'single')

# each rows of X , containing pairs of our features, is assigned a number from 1 to 8 (rows 1 to 8)
labelList = range(1, r+1)  

plt.figure(figsize=(10, 7))
plt.title('Hierarchical Clustering Dendrogram for Correlation and Dissimilarity')
dendrogram(linked,
            orientation='top',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()
```

```python

from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
C = cluster.fit_predict(X)

for i in range(0,r):
    if C[i]==0:
        print(f'Point {i+1} is in class 0' )
    if C[i]==1:
        print(f'Point {i+1} is in class 1' )
```

```python
plt.scatter(X[:,0],X[:,1], c=cluster.labels_, cmap='rainbow')
```
