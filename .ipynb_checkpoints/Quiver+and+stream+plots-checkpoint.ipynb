# Quiver plot with one arrow:

ax.quiver(x_pos, y_pos, x_direct, y_direct)

```python
import numpy as np
import matplotlib.pyplot as plt
# if using a Jupyter notebook, include:
%matplotlib inline

fig, ax = plt.subplots()


x_pos = 0
y_pos = 0
x_direct = 1
y_direct = 1


ax.quiver(x_pos, y_pos, x_direct, y_direct)
ax.set_title('Quiver plot with one arrow')


plt.show()
```

```python
fig, ax = plt.subplots()

x_pos = [0, 0]
y_pos = [0, 0]
x_direct = [1, 0]
y_direct = [1, -1]


ax.quiver(x_pos,y_pos,x_direct,y_direct,
         scale=5)
ax.axis([-1.5, 1.5, -1.5, 1.5])


plt.show()
```

# Use meshgrid to create larger fields

```python
import numpy as np
import matplotlib.pyplot as plt
# if using a Jupyter notebook, include:
%matplotlib inline

x = np.arange(0,2.2,0.2)
y = np.arange(0,2.2,0.2)


X, Y = np.meshgrid(x, y)
u = np.cos(X)*Y
v = np.sin(Y)*y

fig, ax = plt.subplots(figsize=(7,7))
ax.quiver(X,Y,u,v)

ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.axis([-0.2, 2.3, -0.2, 2.3])
ax.set_aspect('equal')


plt.show()
```

$\vec{F} = \frac{x}{5} \ \hat{i} - \frac{y}{5} \ \hat{j}$

```python
x = np.arange(-1,1,0.1)
y = np.arange(-1,1,0.1)

X, Y = np.meshgrid(x, y)
u = np.cos(X)*Y
v = np.sin(Y)*Y


X,Y = np.meshgrid(x,y)


u = X/5
v = -Y/5


fig, ax = plt.subplots(figsize=(9,9))


ax.quiver(X,Y,u,v)


ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.set_aspect('equal')


plt.show()
```

$z = xe^{-x^2-y^2}$

```python
import numpy as np
import matplotlib.pyplot as plt
# if using a Jupyter notebook, include:
%matplotlib inline

x = np.arange(-2,2.2,0.2)
y = np.arange(-2,2.2,0.2)


X, Y = np.meshgrid(x, y)
z = X*np.exp(-X**2 -Y**2)
dx, dy = np.gradient(z)


fig, ax = plt.subplots(figsize=(9,9))


ax.quiver(X,Y,dx,dy)


ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.set_aspect('equal')


plt.show()
```

$\vec{F} = sin(x)cos(y) \ \hat{i} -cos(x)sin(y) \ \hat{j}$

```python
import numpy as np
import matplotlib.pyplot as plt
# if using a Jupyter notebook, include:
%matplotlib inline

x = np.arange(0,2*np.pi+2*np.pi/20,2*np.pi/20)
y = np.arange(0,2*np.pi+2*np.pi/20,2*np.pi/20)


X,Y = np.meshgrid(x,y)


u = np.sin(X)*np.cos(Y)
v = -np.cos(X)*np.sin(Y)


fig, ax = plt.subplots(figsize=(9,9))


ax.quiver(X,Y,u,v)


ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.axis([0,2*np.pi,0,2*np.pi])
ax.set_aspect('equal')


plt.show()
```

# Quiver plots with color:
ax.quiver(x_pos, y_pos, x_direct, y_direct, color)


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

fig, (ax1,ax2,ax3) = plt.subplots(1,3, figsize=(16,16))

# first subplot
x = np.arange(0,2.2,0.2)
y = np.arange(0,2.2,0.2)
X, Y = np.meshgrid(x, y)
u = np.cos(X)*Y
v = np.sin(y)*Y
n = -2
color = np.sqrt(((v-n)/2)**2 + ((u-n)/2)**2)


ax1.quiver(X,Y,u,v,color, alpha=0.8)


ax1.xaxis.set_ticks([])
ax1.yaxis.set_ticks([])
ax1.axis([-0.2, 2.3, -0.2, 2.3])
ax1.set_aspect('equal')
ax1.set_title('meshgrid function')


# second subplot
x = np.arange(-2,2.2,0.2)
y = np.arange(-2,2.2,0.2)
X, Y = np.meshgrid(x, y)
z = X*np.exp(-X**2 -Y**2)
dx, dy = np.gradient(z)
n = -2
color = np.sqrt(((dx-n)/2)**2 + ((dy-n)/2)**2) 


ax2.quiver(X,Y,dx,dy,color)


ax2.xaxis.set_ticks([])
ax2.yaxis.set_ticks([])
ax2.set_aspect('equal')
ax2.set_title('gradient')


# third subplot
x = np.arange(0,2*np.pi+2*np.pi/20,2*np.pi/20)
y = np.arange(0,2*np.pi+2*np.pi/20,2*np.pi/20)
X,Y = np.meshgrid(x,y)
u = np.sin(X)*np.cos(Y)
v = -np.cos(X)*np.sin(Y)
n = -1
color = np.sqrt(((dx-n)/2)**2 + ((dy-n)/2)**2)


ax3.quiver(X,Y,u,v,color)


ax3.xaxis.set_ticks([])
ax3.yaxis.set_ticks([])
ax3.axis([0,2*np.pi,0,2*np.pi])
ax3.set_aspect('equal')
ax3.set_title('four vortices')


# save and show the figure
plt.tight_layout()
fig.savefig('3_quiver_plots.png', dpi=300, bbox_inches='tight')


plt.show()
```

#  Motion fields that I used with Discrete Helmholtz Hodge Decomposition  (2D)

First, Rotational field (notice order of dx, dy in quiver) -   dx, -dy

```python
x = np.arange(-2,2.2,0.2)
y = np.arange(-2,2.2,0.2)
X, Y = np.meshgrid(x, y)
z = np.exp(-X**2 -Y**2)
dx, dy = np.gradient(z)

plt.quiver(X,Y,dx,-dy)
```

Next, convergence towards a point (radial motion) (notice order of dx, dy in quiver) -   dy, dx

```python
x = np.arange(-2,2.2,0.2)
y = np.arange(-2,2.2,0.2)
X, Y = np.meshgrid(x, y)
z = np.exp(-X**2 -Y**2)
dx, dy = np.gradient(z)

plt.quiver(X,Y,dy,dx)
```

Divergence towards a point (radial motion towards a point) (notice order of dx, dy in quiver) -   dy, dx

```python
x = np.arange(-2,2.2,0.2)
y = np.arange(-2,2.2,0.2)
X, Y = np.meshgrid(x, y)
z = np.exp(-X**2 -Y**2)
dx, dy = np.gradient(z)

plt.quiver(X,Y,-dy,-dx)
```

```python
import numpy as np
import matplotlib.pyplot as plt
# if using a Jupyter notebook, include:
%matplotlib inline

x = np.arange(0,10)
y = np.arange(0,10)


X, Y = np.meshgrid(x,y)
u = np.ones((10,10)) # x-component to the right
v = np.zeros((10,10)) # y-component zero


fig, ax = plt.subplots()


ax.streamplot(X,Y,u,v, density = 0.5)
ax.set_title('Stream Plot of Parallel Lines')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# if using a Jupyter notebook, include:
%matplotlib inline

x = np.arange(0,2.2,0.1)
y = np.arange(0,2.2,0.1)


X, Y = np.meshgrid(x, y)
u = np.cos(X)*Y
v = np.sin(y)*Y


fig, ax = plt.subplots()


ax.streamplot(X,Y,u,v, density = 1)
ax.axis([0.5,2.1,0,2])
ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.set_title('Stream Plot of Field Lines')


plt.show()
```

# Streamplots of two charges:

```python
import numpy as np
import matplotlib.pyplot as plt
# if using a Jupyter notebook, include:
%matplotlib inline

x = np.arange(-4,4,0.2)
y = np.arange(-4,4,0.2)


X,Y = np.meshgrid(x,y)
Ex = (X + 1)/((X+1)**2 + Y**2) - (X - 1)/((X-1)**2 + Y**2)
Ey = Y/((X+1)**2 + Y**2) - Y/((X-1)**2 + Y**2)


fig, ax = plt.subplots(figsize=(6,6))


ax.streamplot(X,Y,Ex,Ey)


ax.set_aspect('equal')
ax.plot(-1,0,'-or')
ax.plot(1,0,'-ob')
ax.set_title('Stream Plot of Two Point Charges')


plt.show()
```

# Streamplots for Gaussian functions (circular motion around a point, rotation, convergence towards a point.)

```python
x = np.arange(-2,2.2,0.2)
y = np.arange(-2,2.2,0.2)
X, Y = np.meshgrid(x, y)
z = np.exp(-X**2 -Y**2)
dx, dy = np.gradient(z)

plt.streamplot(X,Y,dx,-dy)  # I had to swap dy for -dy

plt.title('Stream Plot of rotation')

plt.show()
```

```python
x = np.arange(-2,2.2,0.2)
y = np.arange(-2,2.2,0.2)
X, Y = np.meshgrid(x, y)
z = np.exp(-X**2 -Y**2)
dx, dy = np.gradient(z)

plt.streamplot(X,Y,dy,dx)  # I had to swap dy for -dy

plt.title('Stream Plot of convergence')

plt.show()
```

# create a superposition of radial and rotational fields. 

Think of water going down a sink - it is rotating and moving towards the centre. 

```python
x = np.arange(-2,2.2,0.2)
y = np.arange(-2,2.2,0.2)
X, Y = np.meshgrid(x, y)
z = np.exp(-X**2 -Y**2)
dx, dy = np.gradient(z)

plt.streamplot(X,Y,dy+dx,dx-dy)  # I had to swap dy for -dy

plt.title('Stream Plot of convergence')

plt.show()
```

#  Add a small harmonic field (i.e. linear field) 

Put a constant factor in the streamplot

```python


x = np.arange(-2,2.2,0.2)
y = np.arange(-2,2.2,0.2)
X, Y = np.meshgrid(x, y)
z = np.exp(-X**2 -Y**2)
dx, dy = np.gradient(z)

plt.streamplot(X,Y,dy+dx + 0.1 ,dx-dy)  # I had to swap dy for -dy

plt.title('Stream Plot of convergence')

plt.show()
```

```python
# Or plot the above plot as a quiver. 

plt.quiver(X,Y,dy+dx + 0.1 ,dx-dy)
```
